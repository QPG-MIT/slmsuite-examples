Let's still try to get a bit better calibration.
One potential reason for noise in the data is the challenge of exposing the edges of the spots
where certain Zernike terms manifest as perturbations. We can use the high dynamic range (HDR)
feature availible on all `slmsuite` cameras to stitch an image together using multiple exposures.
This allows us to view effectively features below a single level of the camera's bitdepth.

```python
img = cam.get_image(hdr=False)
img_hdr = cam.get_image(hdr=5)

fig, axs = plt.subplots(1, 2, figsize=(20, 10))

cam.plot(img, ax=axs[0], title="No HDR", limits=.25)
plt.clim(0, 1)
cam.plot(img_hdr, ax=axs[1], title="HDR", limits=.25)
plt.clim(0, 1)
```
